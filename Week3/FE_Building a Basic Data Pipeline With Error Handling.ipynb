```python
import pandas as pd
import logging
from pymongo import MongoClient
from pymongo.errors import PyMongoError

# Setup Logging
# Create a logger
logger = logging.getLogger()

# Create handlers
file_handler = logging.FileHandler('C:/Users/Faisal/Desktop/errors/pipeline.log')  # Log file path
file_handler.setLevel(logging.INFO)  # General logs in 'pipeline.log'

# Create a formatter
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')

# Set formatter for handlers
file_handler.setFormatter(formatter)

# Add handlers to the logger
logger.addHandler(file_handler)

# Step 1: Extract Data from CSV
def extract_data(file_path):
    try:
        data = pd.read_csv("https://raw.githubusercontent.com/falawar7/AAI_634O/refs/heads/main/Week3/sales.csv")
        logger.info("Data successfully extracted from CSV file.")
        return data
    except FileNotFoundError as e:
        logger.error(f"File not found: {file_path}")
        raise
    except Exception as e:
        logger.error(f"Error reading CSV file: {e}")
        raise

# Step 2: Transform Data (Calculate Total Revenue)
def transform_data(data):
    try:
        data['total_revenue'] = data['quantity'] * data['price']
        logger.info("Data successfully transformed: total revenue calculated.")
        return data
    except KeyError as e:
        logger.error(f"Missing expected column in data: {e}")
        raise
    except Exception as e:
        logger.error(f"Error transforming data: {e}")
        raise

# Step 3: Load Data into MongoDB
def load_data_to_mongodb(data, db_url, db_name, collection_name):
    try:
        # Connect to MongoDB
        client = MongoClient(db_url)
        db = client[db_name]
        collection = db[collection_name]

        # Log successful connection
        logger.info(f"Successfully connected to MongoDB at {db_url}. Database: {db_name}, Collection: {collection_name}.")

        # Convert the dataframe to a list of dictionaries
        records = data.to_dict(orient='records')

        # Insert the records into the MongoDB collection
        collection.insert_many(records)
        logger.info(f"Data successfully loaded into MongoDB collection '{collection_name}' in database '{db_name}'.")
    except PyMongoError as e:
        logger.error(f"Error connecting to MongoDB or inserting data: {e}")
        raise
    except Exception as e:
        logger.error(f"Unexpected error during MongoDB data load: {e}")
        raise


# Main pipeline function
def run_pipeline(file_path, db_url, db_name, collection_name):
    try:
        # Step 1: Extract
        data = extract_data(file_path)

        # Step 2: Transform
        transformed_data = transform_data(data)

        # Step 3: Load into MongoDB
        load_data_to_mongodb(transformed_data, db_url, db_name, collection_name)

    except Exception as e:
        logger.error(f"Pipeline execution failed: {e}")


# Run the pipeline
if __name__ == '__main__':
    # Define the file path, MongoDB connection parameters
    file_path = 'sales.csv'
    db_url = 'mongodb://localhost:27017/'
    db_name = 'e-commerce_local'  # Updated database name
    collection_name = 'sales_revenue'  # Updated collection name

    # Execute the data pipeline
    run_pipeline(file_path, db_url, db_name, collection_name)
```
